#  Research question1:Is there a relationship between the time of day and departure delays?

# Constructing time_of_day feature

```python
def time_of_day(hour):
    if hour >= 0 and hour < 6:
        return 'night'
    elif hour >= 6 and hour < 12:
        return 'morning'
    elif hour >= 12 and hour < 18:
        return 'afternoon'
    else:
        return 'evening'

df['TIME_OF_DAY'] = df['SCHEDULED_DEPARTURE'].apply(lambda x: time_of_day(x.hour))
```

```python
# Create a box plot of the departure delay by time of day
sns.boxplot(x='TIME_OF_DAY', y='DEPARTURE_DELAY', data=df)
#sns.countplot
# Show the plot
plt.show()
```

```python
# Create a bar plot of the average departure delay by time of day
sns.barplot(x='TIME_OF_DAY', y='DEPARTURE_DELAY', data=df)

# Show the plot
plt.show()
```

# Yes Time_of_day affects the departure_delay,as evening has the biggest average departure delay while night has lowest average

```python
df['TOTAL_DELAY'] = df['DEPARTURE_DELAY'] + df['ARRIVAL_DELAY']
```

# Constructing new features called speed

```python
# Convert the elapsed time from minutes to hours
df['ELAPSED_TIME'] = df['ELAPSED_TIME'] / 60

# Calculate the speed of each flight in miles per hour
df['speed'] = df['DISTANCE'] / df['ELAPSED_TIME']
```

```python
# Group the flights by air time
grouped = df.groupby('AIR_TIME')

# Calculate the mean speed for each group
mean_speeds = grouped['speed'].mean()

# Plot the mean speed for each group
plt.plot(mean_speeds.index, mean_speeds.values)
plt.xlabel('Air time (hours)')
plt.ylabel('Mean speed (miles per hour)')
plt.show()
```
